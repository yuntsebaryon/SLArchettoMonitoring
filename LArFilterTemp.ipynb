```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
fTempTCFilter0 = '/Users/yuntse/data/slarchetto/monitoring/TempTCFilter0_202204201100.csv'
fTempTCFilter1 = '/Users/yuntse/data/slarchetto/monitoring/TempTCFilter1_202204201100.csv'
fTempTCFilter2 = '/Users/yuntse/data/slarchetto/monitoring/TempTCFilter2_202204201100.csv'
fTempTCFilter3 = '/Users/yuntse/data/slarchetto/monitoring/TempTCFilter3_202204201100.csv'
```

```python
# Note in this time era, the temperature was recorded in Kelvin
TempTCFilter0 = pd.read_csv(fTempTCFilter0, header = None, skiprows = 1, names = ['Time', 'Temperature'])
TempTCFilter1 = pd.read_csv(fTempTCFilter1, header = None, skiprows = 1, names = ['Time', 'Temperature'])
TempTCFilter2 = pd.read_csv(fTempTCFilter2, header = None, skiprows = 1, names = ['Time', 'Temperature'])
TempTCFilter3 = pd.read_csv(fTempTCFilter3, header = None, skiprows = 1, names = ['Time', 'Temperature'])
```

```python
TempTCFilter0['Time'] = TempTCFilter0['Time'].transform(lambda t: datetime.strptime(t, '%Y-%m-%d %H:%M:%S.%f'))
TempTCFilter1['Time'] = TempTCFilter1['Time'].transform(lambda t: datetime.strptime(t, '%Y-%m-%d %H:%M:%S.%f'))
TempTCFilter2['Time'] = TempTCFilter2['Time'].transform(lambda t: datetime.strptime(t, '%Y-%m-%d %H:%M:%S.%f'))
TempTCFilter3['Time'] = TempTCFilter3['Time'].transform(lambda t: datetime.strptime(t, '%Y-%m-%d %H:%M:%S.%f'))
```

```python
TempTCFilter0['Time']
```

```python
plt.figure(figsize = (12,8))
plt.plot(TempTCFilter0['Time'], TempTCFilter0['Temperature'] - 273.15, label = 'Filter Top', lw = 3)
plt.plot(TempTCFilter1['Time'], TempTCFilter1['Temperature'] - 273.15, color = 'tab:green', label = 'Filter 2nd', lw = 3)
plt.plot(TempTCFilter2['Time'], TempTCFilter2['Temperature'] - 273.15, color = 'tab:red', label = 'Filter 3rd', lw = 3)
plt.plot(TempTCFilter3['Time'], TempTCFilter3['Temperature'] - 273.15, color = 'tab:orange', label = 'Filter Bottom', lw = 3)


plt.legend(loc = 'lower right', ncol = 1, fontsize = 30)
plt.xticks(fontsize = 26, rotation = 45)
plt.yticks(fontsize = 26)
plt.grid(which = "both", linestyle = '--')
plt.xlabel('Time', fontsize = 26)
plt.ylabel('Temperature ['+u'\u00b0'+'C]', fontsize = 26)

plt.xlim( TempTCFilter0['Time'][0], TempTCFilter0['Time'][-1:])
plt.tight_layout()
plt.savefig('/Users/yuntse/Documents/DUNE/SLArchetto/Monitoring/TempFilter_202204201100.png')
plt.show()
```
